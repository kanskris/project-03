```python
from bs4 import BeautifulSoup as bs
import requests 
import json
import pprint
import pandas as pd
from lxml import html
import requests
pd.options.display.max_rows = 100
pd.options.display.max_columns = 100
```

```python
newURL = 'https://www.nasdaq.com/market-activity/stocks/vz/news-headlines'
response = requests.get(newURL).content
soup = bs(response, 'lxml')
links = soup.body.find_all(class_='quote-news-headlines__item-title')

# https://www.nasdaq.com/articles/walgreens-to-expand-drive-through-coronavirus-testing-sites-2020-04-07
texts = []
links

for l in links:
# print(l.text)
    newLine = l.text
    newLink = "https://www.nasdaq.com/articles/" + newLine.replace(' ','-')
    print(newLink)
    texts.append(newLink)



```

```python
df = pd.DataFrame({'texts':texts})
df.head()
```

```python
def make_clickable(val):
    return '<a href="{}">{}</a>'.format(val,val)

df.style.format(make_clickable)
```

```python
df.to_excel('VZhyperlinkdata.xlsx',index=False)
```
